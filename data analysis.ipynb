```python
# Loading Data

import pandas as pd
data = pd.read_csv('starwars.csv',encoding='latin1')
```

```python
# Settings for all rows and cols
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Print head
data.head()
```

```python
# Changing col names 
data.columns = ['RespondentID','Seen any movie','Fan of franchise?','Movie seen episode 1','Movie seen episode 2',
                'Movie seen episode 3','Movie seen episode 4','Movie seen episode 5','Movie seen episode 6',
                'Movie rank episode 1','Movie rank episode 2','Movie rank episode 3','Movie rank episode 4',
                'Movie rank episode 5','Movie rank episode 6','Han Solo','Luke Skywalker','Princess Leia Organa',
                'Anakin Skywalker','Obi Wan Kenobi','Emperor Palpatine','Darth Vader','Lando Calrissian','Boba Fett',
                'C-3P0','R2 D2','Jar Jar Binks','Padme Amidala','Yoda','Shot first','Familiar with expanded universe?',
                'Fan of expanded universe?','Fan of star trek franchise?','Gender','Age','Income','Education','Location']

# Making the column names consistent
data.columns = data.columns.str.lower()
data.columns = data.columns.str.split(" ")
data.columns = data.columns.str.join("_")
```

```python
data.head()
```

```python
# Checking the data is same or not
data.info()
```

```python
# Checking data types
data.dtypes
```

```python
# Removing null values from 1st column
data = data[pd.notnull(data['respondentid'])]
```

```python
data.head()
```

```python
# Using value counts for each column for sanity checks

data['seen_any_movie'].value_counts()
```

```python
# Stripping extra spaces
data['seen_any_movie'] = data['seen_any_movie'].str.strip()
data['seen_any_movie'].value_counts()
```

```python
data['fan_of_franchise?'].value_counts()
```

```python
# Replacing illegal values
data['fan_of_franchise?'] = data['fan_of_franchise?'].replace("Yess","Yes")
data['fan_of_franchise?'] = data['fan_of_franchise?'].replace("Noo","No")
data['fan_of_franchise?'] = data['fan_of_franchise?'].fillna('No')
data['fan_of_franchise?'].value_counts()
```

```python
# Changing values in columns 3-8
import numpy as np

seen_notseen = {
    
    'seen_notseen_1': {
        data.iloc[0,3]: "Yes",
        np.NaN: "No"
    },

    'seen_notseen_2': {
        data.iloc[0,4]: "Yes",
        np.NaN: "No"
    },

    'seen_notseen_3': {
        data.iloc[0,5]: "Yes",
        np.NaN: "No"
    },
    
    'seen_notseen_4': {
        data.iloc[0,6]: "Yes",
        np.NaN: "No"
    },
    
    'seen_notseen_5': {
        data.iloc[0,7]: "Yes",
        np.NaN: "No"
    },

    'seen_notseen_6': {
        data.iloc[0,8]: "Yes",
        np.NaN: "No"
    },
}


for movie in range(1,7):
    data['movie_seen_episode_' + str(movie)] = data['movie_seen_episode_' + str(movie)].map(seen_notseen['seen_notseen_' + str(movie)])

data.head()
```

```python
# Next 6 columns data type must be float as ranking
data[data.columns[9:15]] = data[data.columns[9:15]].astype(float)
```

```python
# Filling NAs with 0 rank in the 6 columns
for movie in range(1,7):
    data['movie_rank_episode_' + str(movie)] = data['movie_rank_episode_' + str(movie)].fillna(0);
data.head()
```

```python
#Filling the character columns nas with 'Not Voted'

cols_c = ['han_solo', 'luke_skywalker', 'princess_leia_organa',
       'anakin_skywalker', 'obi_wan_kenobi', 'emperor_palpatine',
       'darth_vader', 'lando_calrissian', 'boba_fett', 'c-3p0', 'r2_d2',
       'jar_jar_binks', 'padme_amidala', 'yoda']
for x in cols_c:
    data[x] = data[x].fillna('Not Voted')
data.head()
```

```python
data['gender'].value_counts()
```

```python
data['gender'] = data['gender'].str.strip()
data['gender'] = data['gender'].replace('female','Female')
data['gender'] = data['gender'].replace('male','Male')
data['gender'] = data['gender'].replace('F','Female')
```

```python
data['gender'].value_counts()
```

```python
data['age'].value_counts()
```

```python
# Dropping the column with age = 500 as its impossible
data = data[data['age'] != '500']
```

```python
data['gender'].value_counts()
```

```python
data.isnull().sum()
```

```python
# Dropping columns with so many null values i.e. fan_of_expanded_universe? as it has 972 null values
data=data.drop(['fan_of_expanded_universe?'],axis=1)
```

```python
data.isnull().sum()
```

```python
data['shot_first'].value_counts()
```

```python
data['shot_first'] = data['shot_first'].fillna('Not Known')
```

```python
data['familiar_with_expanded_universe?'].value_counts()
```

```python
data['familiar_with_expanded_universe?'] = data['familiar_with_expanded_universe?'].fillna('No')
```

```python
data['fan_of_star_trek_franchise?'].value_counts()
```

```python
data['fan_of_star_trek_franchise?'] = data['fan_of_star_trek_franchise?'].str.strip()
data['fan_of_star_trek_franchise?'] = data['fan_of_star_trek_franchise?'].replace('yes','Yes')
data['fan_of_star_trek_franchise?'] = data['fan_of_star_trek_franchise?'].replace('no','No')
data['fan_of_star_trek_franchise?'] = data['fan_of_star_trek_franchise?'].replace('Noo','No')
data['fan_of_star_trek_franchise?'] = data['fan_of_star_trek_franchise?'].fillna('No')
```

```python
data['gender'] = data['gender'].fillna('Not Interested in disclosing')
data['gender'].value_counts()
```

```python
for x in ['age','income','education','location']:
    data[x] = data[x].fillna('Not Voted')
```

```python
# Relationship between star war fans and non fans regarding their rankings :
# Fan of Star wars franchise?(Yes/No) VS Rankings of every episode

fans = data[data['fan_of_franchise?'] == 'Yes']
not_fans = data[data['fan_of_franchise?'] == 'No']

agg_data_fans = {}
agg_data_not_fans = {}

cols = ['movie_rank_episode_1','movie_rank_episode_2','movie_rank_episode_3',
       'movie_rank_episode_4','movie_rank_episode_5','movie_rank_episode_6']
av_rank_fans = {}
av_rank_not_fans = {}

for col in cols:   
    av_rank[col] = fans[col].mean()
    av_rank_not_fans[col] = not_fans[col].mean()
    
agg_data_fans['mean'] = av_rank
agg_data_not_fans['mean'] = av_rank_not_fans

rank_fans = sorted(agg_data_fans['mean'].items())
rank_not_fans = sorted(agg_data_not_fans['mean'].items())

idx = np.asarray([i for i in range(len(cols))])

fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)

ax1.bar(idx, [x[1] for x in rank_fans], align='center')
ax1.set_xticklabels(['', 'Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI'])
ax1.set_ylabel('Rank')
ax1.set_title('Average Ranking, by Episode\nSTAR WARS FANS')

ax2.bar(idx, [x[1] for x in rank_not_fans], align='center')
ax2.set_xticklabels(['', 'Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI'])
ax2.set_ylabel('Rank')
ax2.set_title('Average Ranking, by Episode\nSTAR WARS (NOT FANS)')

plt.tight_layout()
plt.show()
```

```python
# Comparison between both (Fans,Not Fans) and rankings
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(2,2,1)

width=0.4
ax.bar(idx, [x[1] for x in rank_fans], align='center',width=width)
ax.bar(idx+width, [x[1] for x in rank_not_fans], align='center',width=width)

ax.set_xticklabels(['','Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI'])
ax.set_ylabel('Rank')
ax.set_title('Average Ranking, by Episode\nFans and Non Fans both')
ax.legend(['Fans','Not Fans'])

plt.tight_layout()
plt.show()
```
